```python
import os
import requests

# Get PDF document path
path = "drive/MyDrive/RAG/human-mutrition-text.pdf"

# Download PDF
if not os.path.exists(path):
    print("[INFO] file doesn't exist, downloading...")

    # Enter URL of PDF
    url = "https://pressbooks.oer.hawaii.edu/humannutrition2/open/download?type=pdf"

    # The local filename to save file
    filename = path

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
      # Write the content of the response to the local file
        with open(filename, "wb") as f:
            f.write(response.content)
        print(f"[INFO] file downloaded successfully as {filename}")
    else:
        print("[ERROR] failed to download file. Status code: (response.status_code)")
else:
    print("[INFO] file already exists")
```

```python
import fitz
from tqdm.auto import tqdm

def text_format(text: str) -> str:

  cleaned_text = text.replace("\n", " ")

  return cleaned_text

def open_pdf(path: str) -> list[dict]:
# Open PDF document
  doc = fitz.open(path)
  page_n_text = []

  for page_number, page in tqdm(enumerate(doc, start=1)):
    text = page.get_text()
    text = text_format(text=text)
    page_n_text.append({"page_number": page_number, "text": text, "char_count": len(text),
                        "word_count": len(text.split(" ")), "sentence_count": len(text.split(". ")),
                        "page_token_count": len(text)/4})
  return page_n_text
```

```python
# Chunking our sentences

sentence_chunk_size = 10

def split_sentences(sentences: list[str],
                    slice_size: int=sentence_chunk_size) -> list[list[str]]:
      return [sentences[i:i + slice_size] for i in range(0, len(sentences), slice_size)]
```

```python
test_list = list(range(35))
split_sentences(test_list)
```

```python
#loop through pages and texts and split sentences into chunks

for item in tqdm(page_n_text):
  item["sentence_chunks"] = split_sentences(item["sentence"],
                                            slice_size=sentence_chunk_size)
  item["num_chunks"] = len(item["sentence_chunks"])
```

#### Splitting each chunk into its own item

```python
import re

page_n_chunks = []
for item in tqdm(page_n_text):
  for chunk in item["sentence_chunks"]:
    chunk_dict = {}
    chunk_dict["page_number"] = item["page_number"]

    #joining sentences into a paragraph
    chunk_text = " ".join(chunk).replace("  ", " ").strip()

    #joined sentence chunk
    chunk_dict["sentence_chunks"] = chunk_text

    #Chunk stats
    chunk_dict["char_count"] = len(chunk_text)
    chunk_dict["word_count"] = len(chunk_text.split(" "))
    chunk_dict["sentence_count"] = len(chunk_text.split(". "))
    chunk_dict["chunk_token_count"] = len(chunk_text)/4

    page_n_chunks.append(chunk_dict)

len(page_n_chunks)
```

```python
from sentence_transformers import SentenceTransformer

embed_model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2",
                                  device="cpu")
```

```python
sentences= ["This framework generates embeddings for each input sentence",
            "Sentences are passed as a list of string.",
            "The quick brown fox jumps over the lazy dog."]

# Sentences are encoded/embedded
embeddings = embed_model.encode(sentences)
embeddings_dict = dict(zip(sentences, embeddings))

# Print the embeddings
for sentence, embedding in embeddings_dict.items():
    print(f"Sentence: {sentence}")
    print(f"Embedding: {embedding}")
    print("")
```

### RAG - Search and Answer

```python
import random

import pandas as pd
import numpy as np
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"

# Import text and embeddings

text_chunks_and_embeddings_df = pd.read_csv("text_chunks_and_embeddings_df.csv")
```

```python
# Convert embedding column to np array
text_chunks_and_embeddings_df["embedding"] = text_chunks_and_embeddings_df["embedding"].apply(lambda x: np.fromstring(x.strip("[]"), sep = " "))
```

```python
text_chunks_and_embeddings_df["embedding"]
```

```python
# Convert embeddings to torch tensor
embeddings = torch.tensor(np.stack(text_chunks_and_embeddings_df["embedding"].to_list(), axis=0), dtype=torch.float32).to(device)
```

```python
embeddings.shape
```

```python
# Convert text and embedding df to list of dict
pages_and_chunks_dict = text_chunks_and_embeddings_df.to_dict(orient="records")
```

```python
pages_and_chunks_dict[:2]
```

```python
from sentence_transformers import SentenceTransformer, util

embed_model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2",
                                  device=device)
```

### Semantic Search pipeline

```python
# 1. Defining Query
query = "Breastfeeding timeline for infants"
print(f"Query: {query}")

# 2. Embed the Query
query_embedding = embed_model.encode(query,
                                      convert_to_tensor=True).to("cuda")

# 3. Get similarity scores
from time import perf_counter as  timer

start_time = timer()
dot_scores = util.dot_score(a= query_embedding, b= embeddings)
end_time = timer()

print(f"[INFO] Time taken to get scores on {len(embeddings)} embeddings: {end_time - start_time} seconds")
```

```python
top_results_dot = torch.topk(dot_scores, k=5)
top_results_dot
```
